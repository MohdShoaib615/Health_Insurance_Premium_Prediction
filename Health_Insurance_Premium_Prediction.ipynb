```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('/content/Health_insurance.csv')
```

```python
plot = px.histogram(df, x='sex', color='smoker')
plot.show()
```

```python
df['sex']=df['sex'].map({'female':0, 'male':1})
df['smoker']=df['smoker'].map({'no':0, 'yes':1})
```

```python
df.head()
```

```python
pie = df['region'].value_counts()
a = pie.index
b = pie.values
fig = px.pie(df, values=b, names=a)
fig.show()
```

```python
df.corr()
```

```python
X = np.array(df[['age', 'sex', 'bmi', 'smoker']])
y = np.array(df['charges'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
pred
```

```python
model.score(X_test,y_test)
```

# **Using the Random Forest, we are acheiving 83% accuracy in our prediction**
